```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)

df_last = pd.read_csv('data/2022-2023.csv')
df_current = pd.read_csv('data/2023-2024.csv')


# Join the two seasons together

df = pd.concat([df_last, df_current], ignore_index=True)
df
```

# Selecting only match related columns. 

- Match Statistics (where available)
- Attendance = Crowd Attendance
- Referee = Match Referee
- HS = Home Team Shots
- AS = Away Team Shots
- HST = Home Team Shots on Target
- AST = Away Team Shots on Target
- HHW = Home Team Hit Woodwork
- AHW = Away Team Hit Woodwork
- HC = Home Team Corners
- AC = Away Team Corners
- HF = Home Team Fouls Committed
- AF = Away Team Fouls Committed
- HFKC = Home Team Free Kicks Conceded
- AFKC = Away Team Free Kicks Conceded
- HO = Home Team Offsides
- AO = Away Team Offsides
- HY = Home Team Yellow Cards
- AY = Away Team Yellow Cards
- HR = Home Team Red Cards
- AR = Away Team Red Cards
- HBP = Home Team Bookings Points (10 = yellow, 25 = red)
- ABP = Away Team Bookings Points (10 = yellow, 25 = red)

```python
df = df[df.columns[:24]].copy()
df.loc[:,'Day'] = df['Date'].apply(lambda x:x.split('/')[0])
df.loc[:,'Month'] = df['Date'].apply(lambda x:x.split('/')[1])
df.loc[:,'Year'] = df['Date'].apply(lambda x:x.split('/')[2])
df.loc[:,'Hour'] = df['Time'].apply(lambda x:x.split(":")[0])
df.loc[:,'Min'] = df['Time'].apply(lambda x:x.split(":")[1])
df = df.drop(columns = ['Date', 'Time'])
df = df.sort_values(by= ['Year', 'Month', 'Day', 'Hour']).reset_index(drop = True)
df
```

```python
# first part we need:
# 1. A row for each match, from the pespective of both the home and away team.
# 2. A column called team_playing
# 3. A column for opponent

df['team_playing'] = df['HomeTeam']
df['opponent'] = df['AwayTeam']

df2 = df.copy()
df2['team_playing'] = df2['AwayTeam']
df2['opponent'] = df2['HomeTeam']
```

```python
# we need:
# Columns describing:
# 1. What the team_playing did. We will use these columns to compute rolling statistics of the team playing.
# 2. Columns describing what the opponent did? Yes because does that represent the team_playing, and how oppositions have done against them right?
#  

df = df[['team_playing', 'opponent', 'HomeTeam', 'AwayTeam', 'FTHG', 'HTHG', 'HS', 'HST', 'HF', 'HC', 'HY', 'HR', 
        'Day', 'Month', 'Year', 'Hour',
        'FTAG', 'HTAG', 'AS', 'AST', 'AF', 'AC', 'AY', 'AR', 'FTR', 'Referee']]

df2 = df2[['team_playing', 'opponent','HomeTeam', 'AwayTeam','FTAG', 'HTAG', 'AS', 'AST', 'AF', 'AC', 'AY', 'AR', 
            'Day', 'Month', 'Year', 'Hour',
            'FTHG', 'HTHG', 'HS', 'HST', 'HF', 'HC', 'HY', 'HR', 'FTR', 'Referee']]


# We have two datasets. Each containing a team_playing column. df is when the team_playing is the home team and df2 is when 
# the team playing is the away team.

# We need to convert these column names from home and way to team_playing stats and opponent stats.
```

```python
# First we change the home team. So we need to rename the columns that represent the home teams to say TP for team playing.


df = df.rename(columns = {'FTHG':'FT_TP_G', 
            'HTHG':'HT_TP_G', 
            'HS':'TP_S', 
            'HST':'TP_ST', 
            'HF':'TP_F',
            'HC': 'TP_C',
            'HY':'TP_Y', 
            'HR':'TP_R',
            'FTAG':'FT_OP_G', 
            'HTAG':'HT_OP_G', 
            'AS':'OP_S', 
            'AST':'OP_ST', 
            'AF':'OP_F',
            'AC': 'OP_C',
            'AY':'OP_Y', 
            'AR':'OP_R'})
df
```

```python
df2 = df2.rename(columns = {'FTHG':'FT_OP_G', 
            'HTHG':'HT_OP_G', 
            'HS':'OP_S', 
            'HST':'OP_ST', 
            'HF':'OP_F',
            'HC': 'OP_C',
            'HY':'OP_Y', 
            'HR':'OP_R',
            'FTAG':'FT_TP_G', 
            'HTAG':'HT_TP_G', 
            'AS':'TP_S', 
            'AST':'TP_ST', 
            'AF':'TP_F',
            'AC': 'TP_C',
            'AY':'TP_Y', 
            'AR':'TP_R'})

df2
```

```python
df_final = pd.concat([df, df2], ignore_index=True)
df_final = df_final.sort_values(by = ['Year', 'Day', 'Month', 'Hour', 'Referee']).reset_index(drop = True)
df_final
```

```python
# create is_home feature.

df_final['is_home'] = (df_final['team_playing'] == df_final['HomeTeam']).astype(int)
df_final
```

```python
df_final
```

```python
# Now we need to create a binary target column that tells us if the team_playing won the match.

df_final['target'] = (((df_final['is_home']==1) & (df_final['FTR']=='H')) | ((df_final['is_home']==0) & (df_final['FTR']=='A')))
df_final
```

```python
df_final['FTR'][0]
```

```python
# Now we have a row for each match from the prespective of both the home and away team. 
# We need to create a binary column to indicate if a team is playing at home or not.

df['is_home'] = (df['HomeTeam'] == df['team_playing']).astype(int)
df
```

```python
# create a target column to indicate if the team_playing won that match or not. 
# 1 - win, 0 - loss and draw.
df['home_win'] = (df['FTHG']>df['FTAG']).astype(int)
df['away_win'] = (df['FTHG']<df['FTAG']).astype(int)
df['draw'] = (df['FTHG']==df['FTAG']).astype(int)
df
```

```python
# target column is. if team_playing won the match then = 1 else 0

df['target'] = (((df['is_home'] == True) & (df['home_win'] == True)) | ((df['is_home'] == False) & (df['away_win'] == True))).astype(int)
# if is_home == 1 and home_win == 1 .apply(lambda x: 1 if (x['is_home']==1 & x['home_win']==1) else 0)then target == 1 
#df = df.drop(columns = ['Referee'])
```

```python
df.sort_values(by = ['Date', 'Hour', 'Referee'], inplace = True)
df = df[['team_playing', 'opponent', 'Hour', 'is_home', 'FTHG', 'FTAG', 'FTR', 'HTHG', 'HTAG', 'HTR', 'HS', 'AS', 'HST', 'AST',
    'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR', 'target']]
df
```
